## Use the pseudo-inverse approach to find a solution to the problem

First, we would like to apply Singular Value Decomposition(SVD) to Matrix $A$, simply use the `svd` function from `numpy.linalg` module

```python
U, S, VT = la.svd(A)
print("U:", U)
print("Singular values: ", S)
print("VT:", VT)
```

Then we would like to compute the pseudo inverse $A^+$ by
<br>
$$
A^+ = V_p\Sigma_p^{-1} U_p^{T}
$$
<br>
<br>
Note that we should suppress the significantly small singular values of $A$ (setting them to zero) when computing $\Sigma$ and $\Sigma^+$, otherwise the reciprocal of which would be very large and break the equation.


```python
# suppress the singular value which close to 0
S[S < 1e-6] = 0
sigma = np.zeros((3, 4))
for i in range(3):
    sigma[i, i] = S[i]
print('checking if the decomposion still holds (U * sigma * VT = A?) after suppressing small singular value:\n')
print('U * sigma * VT : \n', U @ sigma @ VT)

sigma_plus = np.zeros((4, 3))
for i in range(2):
    sigma_plus[i, i] = 1/sigma[i, i]

A_plus = VT.T @ E_plus @ U.T
print('\n A_plus:\n', A_plus)
```

Then we will have our solution $\boldsymbol{x}^+$ by this approach:
$$
\boldsymbol{x}^+ = A^+\boldsymbol{b}

```python
x_plus = A_plus @ b
print('x+ :', x_plus)
```

## A theoretical approach to demonstrate that the solution obtained using the pseudo-inverse is the minimal norm solution

we first define our functional 
 $$ \mathcal{J}\,,\,\mathbb{R}^4 \longrightarrow \mathbb{R} : \mathcal{J}[\boldsymbol{x}]= \boldsymbol{x}^T\boldsymbol{x}$$
 
 with a constraint 
 $$ \mathcal{G}\,,\,\mathbb{R}^4 \longrightarrow \mathbb{R}^4 : \mathcal{G}[\boldsymbol{x}]= A \boldsymbol{x} - \boldsymbol{b} = \boldsymbol{0}$$
 <br>
 <br>
 The necessary condition for the functional $ \mathcal{J}$ to reach its extrema is that when the *Lagrangian* for this system
$$
\mathcal{L}(\boldsymbol{x}, \boldsymbol{\lambda}) :=  \mathcal{J} - \boldsymbol{\lambda}^T \mathcal{G} = \boldsymbol{x}^T\boldsymbol{x} - \boldsymbol{\lambda}^T (A\boldsymbol{x} - \boldsymbol{b})
$$
satisfies 
$$
\left\{
\begin{aligned}
\boldsymbol{0} &=\nabla_{\boldsymbol{x}}\mathcal{L} = 2\boldsymbol{x} - A^T\boldsymbol{\lambda}\\
\boldsymbol{0} &=\nabla_{\boldsymbol{\lambda}}\mathcal{L} = A\boldsymbol{x} - \boldsymbol{b}
\end{aligned}
\right.
$$
<br>
<br>
Also, it's quite obvious to see that $\mathcal{J}$ has no maximum, as we can make  $\boldsymbol{x}$'s norm arbitrarily big. Thus indeed we are finding the minimum of $\mathcal{J}$ under this condition, aka the minimal norm solution to this problem.
 
$$\boldsymbol{0} = 2\boldsymbol{x} - A^T\boldsymbol{\lambda} \implies \boldsymbol{x} = \frac{1}{2} A^T\boldsymbol{\lambda} $$
substitute into the second equation we have
$$
\begin{aligned}
A \left(\frac{1}{2} A^T\boldsymbol{\lambda}\right) &= \boldsymbol{b}\\
A A^T\boldsymbol{\lambda} &=  2\,\boldsymbol{b}
\end{aligned}
$$

Recall in ACSE-7 lecture 1, we have the following properties for the pseudo inverse:

$$
\begin{align*}
AA^+A &= A\\
A^+AA^+ &= A^+\\
(A^+A)^T &= A^+A\\
(AA^+)^T &= AA^+\\
\end{align*}
$$

Thus, we can see that:
$$
\begin{align*}
A^T &= \left(A A^+ A\right)^T\\
&= \left(A^+ A\right)^T \, A^T\\
&= A^+A \, A^T
\end{align*}
$$

left multiply $A^+$ on both sides for the equation 
$$
A A^T\boldsymbol{\lambda} =  2\,\boldsymbol{b}
$$
we have
$$
\begin{aligned}
A^+ A A^T \boldsymbol{\lambda} &=  A^+ 2\,\boldsymbol{b}\\
\left(A^+ \, A A^T \right) \boldsymbol{\lambda} &=  2\, A^+ \boldsymbol{b}\\
A^T \boldsymbol{\lambda} &=  2\, A^+ \boldsymbol{b}
\end{aligned}
$$

also we have $\boldsymbol{x} = \frac{1}{2} A^T\boldsymbol{\lambda} $
, plug the above result in, we have 
$$
\begin{aligned}
\boldsymbol{x} &= \frac{1}{2} A^T\boldsymbol{\lambda}\\
&= \frac{1}{2} \cdot 2\, A^+ \boldsymbol{b}\\
&= A^+ \boldsymbol{b}
\end{aligned}
$$
Thus, we can see, if the minimal norm solution to this problem exists, it should be the solution
$$
\boldsymbol{x}^+ = A^+\boldsymbol{b}
$$
which is computed using the pseudo-inverse approach.

## Find the solution that minimises the first two entries of $\boldsymbol{x}$

As we know, any valid solution for the equation can be write as 
$$\boldsymbol{x} = \boldsymbol{x}^+  + t \cdot \boldsymbol{n_1} + v \cdot \boldsymbol{n_2}$$ 
with $t, v \in \mathbb{R}$,
<br>
<br>
As we only focus to minimize $x_1^2 + x_2^2$, we can just cut off the last two dimensions, reducing dimension for the problem.

Thus, we can just represent our solution with the first two entries $\boldsymbol{x}' = (x_1, x_2)$ with
$$
\boldsymbol{x}' = \boldsymbol{x}^{+ \, '}  + t \cdot \boldsymbol{n_1}' + v \cdot \boldsymbol{n_2}'
$$

where the vectors with $'$ represents a sub-vector only taking its first two entries.

Define $\boldsymbol{x}^{+ \, '}$:

```python
x_plus_sub = np.array([x_plus[0], x_plus[1]])
print("x+':", x_plus_sub)
```

Also 
$$
t \cdot \boldsymbol{n_1}' + v \cdot \boldsymbol{n_2}' 
$$
can be write in a Matrix form 
$$
B \, \boldsymbol{\xi}
$$
<br>
where
$$
B = 
\begin{pmatrix}
-1 & 4 \\
-2 & -1  
\end{pmatrix}
\,\,,\,\,
\boldsymbol{\xi} =
\begin{pmatrix}
t\\
v
\end{pmatrix}
$$

```python
B = np.array([[-1, 4], [-2, -1]])
```

If we want to minimize 
$$
\begin{aligned}
f(\boldsymbol{\xi}) &= {\boldsymbol{x}'}^T \boldsymbol{x}'\\
&= \left( B \, \boldsymbol{\xi} + \boldsymbol{x}^{+ \, '}\right)^T \left( B \, \boldsymbol{\xi} + \boldsymbol{x}^{+ \, '}\right)\\
&= \boldsymbol{\xi}^ T B^T  B \, \boldsymbol{\xi} + 2 \, {\boldsymbol{x}^{+ \, '}}^T \, B \, \boldsymbol{\xi} \,+\, {\boldsymbol{x}^{+ \, '}}^T \boldsymbol{x}^{+ \, '}
\end{aligned}
$$

we should borrow some knowledge from lecture 2 (Unconstrained Optimisation).
First check if $B^T B$ is SPD:

```python
BTB = B.T @ B
la.eig(BTB)[0]
```

Yes it is, as both of its eigenvalues are positive.

Taking derivative of $f$ w.r.t $\boldsymbol{\xi}$, we have
$$
f'(\boldsymbol{\xi}) = 2 B^T  B \, \boldsymbol{\xi} + 2 \, {\boldsymbol{x}^{+ \, '}}^T \, B
$$
<br>
<br>
As $B^T B$ is SPD, a global minimum is guaranteed when $f'(\boldsymbol{\xi}) = \boldsymbol{0}$.

That is when 
$$
B^T  B \, \boldsymbol{\xi} = - \, {\boldsymbol{x}^{+ \, '}}^T \, B
$$

Solve for $\boldsymbol{\xi}$ under this condition:

```python
xi_optimal = la.solve(BTB, -b0.T @ B)
print("The t and v (when x1^2 + x2^2 is the smallest):", xi_optimal)
```

Again, we use the contour plot to act as an evidence for our solution:

```python
dx = 0.1
x = np.arange(-7.5, 7.5, dx)
y = np.arange(-7.5, 7.5, dx)
X, Y = np.meshgrid(x, y)

f = np.zeros(X.shape)
for i in range(X.shape[0]):
    for j in range(X.shape[1]):    
        f[i, j] = (x_plus + X[i, j] * n1  + Y[i, j] * n2)[0]**2 + (x_plus + X[i, j] * n1  + Y[i, j] * n2)[1]**2

fig = plt.figure(figsize=(5, 5))
ax1 = fig.add_subplot(111)

cs = ax1.contour(X, Y, f, 30)
ax1.clabel(cs, inline=1, fontsize=10)
ax1.set_title('Contour plot of x1^2 + x2^2 and the minima location')
ax1.set_xlabel('$t$', fontsize=12)
ax1.set_ylabel('$v$', fontsize=12)

# adding up n1 and n2 to x_plus with the coefficients computed on the cell above
ax1.plot(xi_optimal[0], xi_optimal[1] ,'r*')
```
